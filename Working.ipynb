```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

 School ranking file - rankings by state


```python
school_DF=pd.read_csv('./Data/school_ranking.csv')
```

Copy of original DF is made - so if there are mistakes, I can always go back to original.  Working copy is called 'working_schoolDF'

```python
working_schoolDF=school_DF.copy()
```

iloc command used to isolate only the columns for state name and overall ranking

```python
working_schoolDF=working_schoolDF.iloc[:,0:2]
display(working_schoolDF)
```

Mental Health data - rankings of mental health variables by state

```python
mental_DF=pd.read_csv('./Data/mental_health.csv')
display(mental_DF)
```

Need to isolate all rows columns=="By State" and columns=="Symptoms of Anxiety Disorder or Depressive Disorder" only, and the columns need to be shown are "Indicator", "Group","Subgroup", "Time Period Label", and "Value"
Start with extracting only those columns, then only choose certain categories 

```python
working_mentalDF=mental_DF.copy()
working_mentalDF=working_mentalDF[["Indicator","Group","Subgroup","Time Period Start Date","Time Period End Date","Value"]]
working_mentalDF=working_mentalDF[(working_mentalDF["Group"]=="By State") & (working_mentalDF["Indicator"]=="Symptoms of Anxiety Disorder or Depressive Disorder")]
display(working_mentalDF)
```

Since the date columns are of type 'object', need to convert to type 'datetime'

```python
working_mentalDF["Time Period Start Date"]=pd.to_datetime(working_mentalDF["Time Period Start Date"])
working_mentalDF["Time Period End Date"]=pd.to_datetime(working_mentalDF["Time Period End Date"])
display(working_mentalDF)
```

Checking the type of columns "Time Period Start Date" and "Time Period End Date"

```python
display(working_mentalDF.info())
```

Need to extract the year only from the dates so can perform arithmetic comparison accordingly

```python
working_mentalDF["Time Period Start Date"]=working_mentalDF["Time Period Start Date"].dt.year
working_mentalDF["Time Period End Date"]=working_mentalDF["Time Period End Date"].dt.year
display(working_mentalDF)
```

To isolate results only for year 2023

```python
working_mentalDF=working_mentalDF[["Subgroup","Time Period Start Date","Time Period End Date","Value"]]
working_mentalDF=working_mentalDF[(working_mentalDF["Time Period Start Date"]==2023) & (working_mentalDF["Time Period End Date"]==2023)]
display(working_mentalDF)
```

Drop the columns "Time Period Start Date" and "Time Period End Date"

```python
working_mentalDF=working_mentalDF.drop(columns=["Time Period Start Date","Time Period End Date"])
working_mentalDF=working_mentalDF.rename(columns={"Subgroup":"State"})
display(working_mentalDF)
```

Get the average of value, grouped by state
Drop "District of Columbia" because there are only 50 states for the other DF (school ranking)

```python
working_mentalDF=working_mentalDF.groupby("State").mean("Value")
working_mentalDF=working_mentalDF.drop(["District of Columbia"])
display(working_mentalDF)
```

Need to rank the value in ascending order because value reflects % of respondents displaying anxiety or depression symptoms

```python
working_mentalDF["Mental Health Ranking"]=working_mentalDF["Value"].rank()
display(working_mentalDF)
```

Sort the rank in ascending order

```python
working_mentalDF=pd.DataFrame(working_mentalDF["Mental Health Ranking"].sort_values())
display(working_mentalDF)
```

Need to join the 2 DF's (school and mental rankings) based on the column state

```python
combined_df1=pd.merge(working_schoolDF,working_mentalDF,left_on="State",right_on="State")
display(combined_df1)
```

Clean up the Dataframe:  
1. Rearrange the order of the columns
2. Rename the columns to be more descriptive
3. Change the ranking numbers in Mental Health to integers

```python
combined_df1=combined_df1.reindex(columns=["State","RANK","Mental Health Ranking"])
combined_df1=combined_df1.rename(columns={"RANK":"School Ranking"})
combined_df1["Mental Health Ranking"]=combined_df1["Mental Health Ranking"].astype(int)
display(combined_df1)
```

To read the income file and make a working copy

```python
income_df=pd.read_csv('./Data/income.csv')
working_incomedf=income_df.copy()
display(working_incomedf)
```

To pick up only certain columns:  "AREA_TITLE", "AREA_TYPE", "OCC_TITLE", "A_MEAN"
Only need by state (Area Title) so Area Type=2 and by OCC Title = All Occupations 

```python
working_incomedf=working_incomedf[["AREA_TITLE","AREA_TYPE","OCC_TITLE","A_MEAN"]]
working_incomedf=working_incomedf[(working_incomedf["AREA_TYPE"]==2) & (working_incomedf["OCC_TITLE"]=="All Occupations")]
display(working_incomedf)
```

Delete the middle 2 columns (Area Type and Occ Title)
Change Column names to State and Mean Income

```python
working_incomedf=working_incomedf.drop(columns=["AREA_TYPE","OCC_TITLE"])
working_incomedf=working_incomedf.rename(columns={"AREA_TITLE":"State","A_MEAN":"Mean Income"})
display(working_incomedf)
```

Delete the row 'District of Columbia' to include only 50 states

```python
working_incomedf=working_incomedf.drop(index=177509)
display(working_incomedf)
```

Rank the level of mean income, in descending order, in integer format (not float)

```python
working_incomedf["Income Ranking"]=working_incomedf["Mean Income"].rank(ascending=False).astype(int)
display(working_incomedf)
```

Merge this dataframe with the 1st combined dataframe (school and mental health rankings)

```python
combined_df2=pd.merge(combined_df1,working_incomedf,left_on="State",right_on="State")
display(combined_df2)
```

To reorder the last 2 columns so Mean Income appears last

```python
combined_df2=combined_df2.reindex(columns=["State","School Ranking","Mental Health Ranking","Income Ranking","Mean Income"])
display(combined_df2)
```

Clean up dataframe to check for quick correlations - get rid of "Median Income" to focus interest on rankings and get rid of "State" because it is not an index column

```python
combined_df3=combined_df2.drop(["Mean Income","State"], axis=1)
combined_df3
```

Quick check for correlations

```python
combined_df3.corr()
```

This correlation table shows that there is pretty good correlation between all 3 variables
